```python
ds = dataset.iloc[:,:-6]
```

Creation de colonnes de compartimentation des retards pour dé corréler les colonnes

```python
ds['Petit Retard'] = ds['Nombre trains en retard > 15min'] - ds['Nombre trains en retard > 30min']
ds['Moyen Retard'] = ds['Nombre trains en retard > 30min'] - ds['Nombre trains en retard > 60min']
ds['Grand Retard'] = ds['Nombre trains en retard > 60min']
ds = ds.drop('Nombre trains en retard > 15min',axis=1)
ds = ds.drop('Nombre trains en retard > 30min',axis=1)
ds = ds.drop('Nombre trains en retard > 60min',axis=1)
```

```python
ds.boxplot(rot=90)
```

Suppression des trajets où la durée moyenne de trajet est de 0 minute et aussi celui de 800 minutes.

Ainsi que ceux dont le retard est négatif.

```python
ds = ds.drop(ds[ds['Durée moyenne du trajet']==0].index)
ds = ds.drop(ds[ds['Retard moyen de tous les trains au départ'] < 0 ].index)
ds = ds.drop(ds[ds['Retard moyen des trains en retard à l\'arrivée']<0].index)
ds = ds.drop(ds[ds['Retard moyen de tous les trains à l\'arrivée']<0].index)
ds = ds.drop(ds[ds['Retard moyen trains en retard > 15 (si liaison concurrencée par vol)']<0].index)
```

Suppression de lignes incohérentes (nombre de retard > 15min devrait être l'addition de ceux >30min et de ceux entre 15 et 30 min -> et donc Petit Retard ne peux être négatif)

```python
ds = ds.drop(ds[ds['Petit Retard']<0].index)
```
