```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.model_selection import KFold
```

```python
df = pd.read_csv('train.csv', index_col=0)
train_x = df[["Energy","Danceability","Valence","Acousticness","Speechiness","Popularity"]]
train_y = df[["Rock","Pop","Standards","Metal","Indie","Cabaret","Soul","Wave","Invasion","Hip-hop"]]
```

```python
def classifier(x_train,y_train,x_test,y_test):
    l = []
    for names in train_y.columns:
        gnb = GaussianNB()
        gnb.fit(x_train,y_train[[names]])
        y_pred = gnb.predict(x_test)
        if names == "Rock":
            print(y_pred)
        l.append(metrics.accuracy_score(y_test[[names]],y_pred)*100)
    return l
```

```python
def mean_list(lists):
    brute_score = [0 for _ in range(len(lists))]
    for l in lists:
        for i in range(len(l)):
            brute_score[i] += l[i]
    for i in range(len(brute_score)):
        brute_score[i] = brute_score[i] / 10
    return brute_score
        
```

```python
def cross_validation():
    num_splits = 10
    folds = KFold(n_splits=num_splits)
    lists = []
    names = train_y.columns
    sum_score = 0

    for train_index, test_index in folds.split(train_x):
        X_train, X_test = train_x.iloc[train_index], train_x.iloc[test_index]
        y_train, y_test = train_y.iloc[train_index], train_y.iloc[test_index]
        lists.append(classifier(X_train, y_train, X_test, y_test))
    
    for i in range(len(lists)):
        sum_score += mean_list(lists)[i] * train_y.value_counts().iloc[i]
        print(names[i],mean_list(lists)[i])

    print(sum_score/len(train_y))

cross_validation()
```
